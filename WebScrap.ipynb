# Teste a partir da Primeira Página

```python
# 

import time
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys # Necessário para enviar caracteres especiais no comando .send_keys()
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Definir site
Site = "https://nvd.nist.gov/vuln/search"

# Abrir o site
driver.get(Site)


### Funções
    
def  Receba(Caminho):    
    '''
    Retorna uma lista com o texto dos elementos encontrados no caminho(xpath) 
    
    '''
    Elemento = driver.find_elements(By.XPATH , Caminho) # contem "Elemento" Tabela resultado da pesquisa
    Lista = []
    for item in Elemento:
        Lista.append(item.text)
    return Lista

def ProximaPagina():  
    '''
    Idendifica existencia de página adicional e segue para próxima
    
    '''
    Navegador = driver.find_elements(By.XPATH, '//*[@id="row"]/div/div/nav/ul/li/a')
    print(len(Navegador))    
    Saida = False
    time.sleep(2) 
    for item in Navegador:
        if item.text == '>':
            print('Abrindo próxima página...')
            item.click()
            Saida = True
            break
        else:
            Saida = False
            
    return Saida    

### Declaração de variáveis (Listas)
lstCVE = []
lstCVE_v20 = []
lstCVE_v31 = []
lstDescr = []
lstPubli = []
sistema = []
link = []

### Declaração de variáveis (Xpath)
XpathCVE_ID = '//*/tbody/tr/th/strong/a'
XpathCVE_v20 = '//*/tbody/tr/td[2]/span[2]'
XpathCVE_v31 = '//*/tbody/tr/td[2]/span[1]'
XpathDescr = '//*/tbody/tr/td[1]/p'
XpathPubli = '//*/table/tbody/tr/td[1]/span'

# Ativar (radio button) busca avançada
AdvavancedSearch = driver.find_element(value='SearchTypeAdvanced')
AdvavancedSearch.click()

# Keyword Search //*[@id="Keywords"]
Keyword = driver.find_element(By.XPATH, '//*[@id="Keywords"]')
Keyword.clear()

# Encontrar os elementos Data inicial e Data final
StartDate = driver.find_element(value='published-start-date')
StartDate.clear()

EndDate = driver.find_element(value='published-end-date')#find_element_by_name('pub_end_date')
EndDate.clear()

# inserir nos elementos os valores de Data inicial e Data final
#Pesquisado = 'Nimbus'
Pesquisado = '' #<<<<<<<<<<<<<<<<Pesquisa
Keyword.send_keys(Pesquisado)
StartDate.send_keys('01012022')
EndDate.send_keys('05012022')

# Clicar no botão "Search"
SubmitSearch = driver.find_element(value='vuln-search-submit')
SubmitSearch.click()
time.sleep(2)

# Encontrar lista de vulnerabilidades
    # Colunas: CVE - Descrição - Pontuação 

UltimaPagina = False        
while not(UltimaPagina):
    lstCVE += Receba(XpathCVE_ID)
    lstCVE_v20 += Receba(XpathCVE_v20)
    lstCVE_v31 += Receba(XpathCVE_v31)
    lstDescr += Receba(XpathDescr)
    lstPubli += Receba(XpathPubli)
    
    UltimaPagina = not(ProximaPagina())
    #print(UltimaPagina)
print('Esta é a última página') 

for i in lstCVE:
    link.append('https://nvd.nist.gov/vuln/detail/'+i)
    sistema.append(Pesquisado)
    
Tabela={}
Tabela.update({'CVE_ID': lstCVE})
Tabela.update({'CVE_v20': lstCVE_v20})
Tabela.update({'CVE_v31': lstCVE_v31})
Tabela.update({'Descrição': lstDescr})
Tabela.update({'Publicação': lstPubli})    
Tabela.update({'Link':link})
Tabela.update({'Sistema':sistema})
'''
Tabela2 = {} 
for i in range(len(Tabela['CVE_ID'])):
    Tabela2.update({'CVE_ID_'+str(i):Tabela['CVE_ID'][i], 'CVE_v31_'+str(i):Tabela['CVE_v31'][i], 'CVE_v20_'+str(i):Tabela['CVE_v20'][i], 'Descrição_'+str(i): Tabela['Descrição'][i]})
    
'''

print(len(Tabela['Publicação']), 'iten(s) na tabela')
print("Conteúdo:")
print(Tabela['Publicação'])

# encerrar o navegador
time.sleep(5)
driver.quit()

# Fechar a aba do navegador
#driver.close()
```

```python
import time

from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys # Necessário para enviar caracteres especiais no comando .send_keys()
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Definir site
Site = 'https://nvd.nist.gov/vuln/search/results?form_type=Advanced&results_type=overview&query=Cyclos&search_type=all&isCpeNameSearch=false&pub_start_date=01%2F01%2F2022&pub_end_date=05%2F01%2F2022'

# Abrir o site
driver.get(Site)
time.sleep(2)
# Encontrar lista de vulnerabilidades
    # Colunas: CVE - Descrição - Pontuação

TabelaDeVulnerabilidades = driver.find_element(By.TAG_NAME , 'tbody') # contem "Elemento" Tabela resultado da pesquisa

ListaDeVulnerabilidades = TabelaDeVulnerabilidades.find_elements(By.TAG_NAME, 'tr') # Contem Lista com Elementos, cada  item da lista é ima linha completa


desc = []
pont = []
cve = []
Pesquisa = {}
Tabela = {}

for i in range(len(ListaDeVulnerabilidades)):
    Tabela.update({i:ListaDeVulnerabilidades[i].text.split('\n')})
    cve.append(Tabela[i][0])
    desc.append(Tabela[i][1])
    pont.append(Tabela[i][3])
    
Tabela.update({'CVE':cve})
Tabela.update({'Descrição':desc})
Tabela.update({'Pontuação':pont})

for item in  Tabela['CVE']:
    print(item,'\n')
        
    
    
    
# encerrar o navegador
time.sleep(15)
driver.quit()

# Fechar a aba do navegador
#driver.close()
```
